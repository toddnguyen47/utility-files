```python
# /----------------------------------------------------------\
# region CONFIG
_FOLDER_PATH = Path(r"Z:\GoogleDrive\Books\light-novels\Secrets of the Silent Witch")
_IGNORE_STR = "Secrets of the Silent Witch"
_EXTENSION = ".epub"
# endregion CONFIG
# \----------------------------------------------------------/

_RE_IGNORE = re.compile(_IGNORE_STR.lower())
m1 = Main()
m1.main()
```

```python
from pathlib import Path
import re
import os

class Data:
    """A data class"""
    old_name: str
    tokens: list[str]

    def __init__(self, old_name: str, tokens: list[str]):
        self.old_name = old_name
        self.tokens = tokens

    def __str__(self):
        return f'{self.old_name}; {"-".join(self.tokens)}'

class Main:
    """main class"""

    def main(self):
        data_list: list[Data] = []

        for root, dirs, files in os.walk(_FOLDER_PATH):
            for file in files:
                full_path = Path(root, file)
                if full_path.suffix.lower() == _EXTENSION.lower():
                    tokens = self._get_new_name_tokens(full_path)
                    d1 = Data(full_path, tokens)
                    data_list.append(d1)
        
        longest_len = 0
        for data in data_list:
            l1 = len(data.tokens)
            if l1 > longest_len:
                longest_len = l1
        
        for data in data_list:
            l1 = len(data.tokens)
            while l1 < longest_len:
                data.tokens.append(f'{0:0<2}')
                l1 += 1
            # rename now
            parent = data.old_name.parent
            name = data.old_name.name
            shortened_name = name.replace(" ", "")[:10].upper()
            new_name = f'{shortened_name}-{"-".join(data.tokens)}'
            new_name = f'{new_name}_{name}'
            new_full_path = Path(parent, new_name)
            print(new_full_path)
            os.rename(data.old_name, new_full_path)
    
    def _get_new_name_tokens(self, file_name: Path) -> (list[str]):
        """Get new name tokens. Returns a list of tokens"""
        parent = file_name.parent
        name = file_name.name
        name_copy = str(name)
        number_l1 = []
    
        # ignore certain parts of the string
        lower_file_name = name.lower()
        a = _RE_IGNORE.search(lower_file_name)
        if a:
            name_copy = name_copy[a.end():]
        
        cur_index, start_index, end_index = 0, 0, 0
        len_name = len(name_copy)
        while cur_index < len_name:
            c = name_copy[cur_index]
            if c.isdigit():
                start_index = cur_index
                end_index = cur_index
                while end_index < len_name and name_copy[end_index].isdigit():
                    end_index += 1
                token = f'{name_copy[start_index:end_index]:0>2}'
                number_l1.append(token)
                # have to subtract 1 to account for adding 1 at the end of each loop
                cur_index = end_index - 1
            cur_index += 1
        return number_l1
```
